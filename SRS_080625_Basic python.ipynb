```python
marks=int(input("Enter your Marks: "))
Grade = 0         #initiate a variable
Status = "Null"     #initiate a variable
if (marks>=0) and (marks<=100):
  if marks>90:
    Grade = "O"
    Status = "Distinction"
  elif marks>80:
    Grade="A"
    Status = "First Class"
  elif marks>65:
    Grade = "B"
    Status = "Second Class"
  elif marks >45:
    Grade = "C"
    Status = "Third Class"
  elif marks >=36:
    Grade = "D"
    Status = "Passed"
  else:
    Grade = "F"
    Status = "Failed"
else:
  print("Invalid Marks")

print(f"Your Grade is {Grade} and your Status is {Status}")
```
