#Visualizations in Python

```python
#install and then import matplotlib
import matplotlib.pyplot as plt
```

```python
dir(plt)
```

```python
#matplotlib to render plots in the notebook
%matplotlib inline
```

```python
x=[1,2,3,4,5,6,7,8,9]
```

```python
y=[]                      #creating variable y as square value of x variable
for i in x:
  y.append(i**2)
```

```python
y
```

```python
y = [i**2 for i in x]   #same as above in simpler way
y
```

```python
plt.plot(x,y)   #plot plots a line chart
```

```python
fig = plt.figure(figsize=(15,5))       #storing a figure in the variable fig

plt.plot(x,y)                     #generate the line plot
plt.ylim(1,100)                #putting limits for y-axis
plt.xlim(1,10)              #putting limits for x-axis
plt.xlabel('Numbers')       #labelling x-axis
plt.ylabel('squares')        #labelling y-axis
plt.title('Line Plot')       #gives a simple title just above the plot
plt.suptitle('Sales comparison', size=20, y =1.03)      #gives the super title with font size and line space between title and super title
plt.show()
```
